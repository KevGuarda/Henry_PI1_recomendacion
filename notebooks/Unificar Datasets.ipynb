```python
import pandas as pd
import numpy as np
import ast
import json
```

```python
#Cargamos la base de datos 'movies_dataset_procesado'
df_mo = pd.read_csv("C:/Users/guard/OneDrive/Desktop/Henry Data Science/Proyecto MLOps/datasets/movies_dataset_procesado.csv", sep=",")

df_mo.head(2)
```

```python
#Cargamos la base de datos 'credits_dataset_procesado'
df_cr = pd.read_csv("C:/Users/guard/OneDrive/Desktop/Henry Data Science/Proyecto MLOps/datasets/credits_dataset_procesado.csv", sep=",")

df_cr.head(2)
```

```python
# Realizar la unión de los dos datasets basándonos en la columna 'id'
df_merged = pd.merge(df_mo, df_cr, on='id', how='left')
```

```python
# Asegurar de que la columna 'release_year' existe y está en el formato correcto
if 'release_year' in df_merged.columns:
    df_merged['release_year'] = df_merged['release_year'].fillna(0).astype(int)

# Redondear las columnas 'popularity' y 'return' a dos dígitos decimales
columns_to_round = ['popularity', 'return']
df_merged[columns_to_round] = df_merged[columns_to_round].round(2)
```

```python
df_merged.head(2)
```

```python
# Revisar dataset resultante en caso de que haya nulos al unir ambos datasets
df_merged.info()
```

```python
# Eliminar las filas que tengan faltantes en la columna 'title'
df_merged = df_merged.dropna(subset=['title'])

# Reemplazar los valores nulos en las columnas float por 0
float_cols = df_merged.select_dtypes(include=['float']).columns
df_merged[float_cols] = df_merged[float_cols].fillna(0)

# Reemplazar los valores nulos en las columnas object por 'Sin Datos'
object_cols = df_merged.select_dtypes(include=['object']).columns
df_merged[object_cols] = df_merged[object_cols].fillna('Sin Datos')
```

```python
df_merged.info()
```

```python
# Exportar el DataFrame a un archivo CSV
df_merged.to_csv('dataset_unificado.csv', index=False)  # index=True incluye el índice en el archivo CSV
```

Vamos a reducir el dataset porque la matriz resultado del Score de similaridad a partir de los títulos queda muy grande y por lo tanto al momento de cargar el archivo en Render este no termina de hacer el deploy, posiblemente cambiando la dimensionalidad del dataset podremos resolver este inconveniente.

```python
# Asegurar de que la columna 'release_date' está en formato de fecha
df_merged['release_date'] = pd.to_datetime(df_merged['release_date'], errors='coerce')

# Eliminar filas con fechas inválidas
df_merged = df_merged.dropna(subset=['release_date'])

# Ordenar el DataFrame por 'release_date' en orden descendente y seleccionar las 25000 películas más recientes
df_filtered = df_merged.sort_values(by='release_date', ascending=False).head(25000)

# Guardar el DataFrame filtrado en un nuevo archivo CSV
df_filtered.to_csv("C:/Users/guard/OneDrive/Desktop/Henry Data Science/Proyecto MLOps/datasets/dataset_filtrado_25000.csv", index=False)

print("Dataset reducido a las 5000 películas más recientes y guardado como 'dataset_filtrado_25000.csv'")
```
